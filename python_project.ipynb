### Child Nutrietion Calculater

```python
#taking general data of any kid
n = input("name of the child: ")
a = int(input('age: '))
g = input('gender: ')
h = float(input('hiegth in inches: '))
w = float(input('weight in  pounds: '))

#calculates the bmi
bmi = w/(h**2)*703
print("BMI of the child is:",bmi)

#checks the status of body health
if bmi<=16:
    print("severly underweight")
elif bmi>=16.0 and bmi<18.5:
    print("underweight")
elif bmi>=18.5 and bmi<25:
    print("healthy")
elif bmi>=25 and bmi<30:
    print("overweight")
elif bmi>=30:
    print("obese")


print("minimum calorie requirement")
if a<2:
    print(800)
if a>=2 and a<4:
    print(1400)
if a>=4 and a<8:
    print(1800)
 

#calculates total calories and compares with daily need

print("enter what child ate, press 0 in food and quatity field to submit")
emptydict = {}

while True:
   x = str(input("enter food: "))
   y = float(input("enter quantity in grams: "))
   emptydict.update({x:y})
   if x=="0":
        break

 # dictionary for calorie data for distict sources of food       
caldata = {"milk":1,"egg":1.55,"rice":1.30,"lentils":1.13,"vegetable":0.85,"meat":1.43}       



#calculate total calories
total = 0
for i in caldata:
    if i in emptydict:
        tc = (caldata.get(i)*emptydict.get(i))
        total = total+tc

        
        
#compares and prints the status of nourishment on the basis of total calorie consumption
print("total calorie consumption is "+str(total)+" your child is")
if a<2 and total<800:
    print("undernourished")
    
elif (a>=2 or a<4) and total<1400:
    print("undernourished")
    
elif (a>=4 or a<8) and total<1800:
    print("undernourished")
else:
    print("nourished")
        

```
